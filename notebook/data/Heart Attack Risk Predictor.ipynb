#Loading Data Set and Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('heart_attack_prediction_dataset.csv')
```

# EDA

```python
#droping columns which don't add value to Prediction ['Patient ID', 'Hemisphere', 'Country', 'Continent']
df.drop(['Patient ID', 'Hemisphere', 'Country', 'Continent'], inplace = True, axis= 1)
```

```python
#Spliting the Blood Pressure Column into two Columns['Systolic BP', 'Diastolic BP'] and removing Blood Pressure Column
df[['Systolic BP', 'Diastolic BP']] = df['Blood Pressure'].str.split('/', expand=True)
df.drop(['Blood Pressure'], inplace = True, axis = 1)
df[['Systolic BP', 'Diastolic BP']] = df[['Systolic BP', 'Diastolic BP']].astype(int)
```

```python
#Ordinal Encoding for Diet Column
from sklearn.preprocessing import OrdinalEncoder
Oencoder = OrdinalEncoder(categories= [['Unhealthy', 'Average', 'Healthy']])
df['Diet'] = Oencoder.fit_transform(df[['Diet']])
df['Diet'] = df['Diet'].astype(int)
```

```python
#Maping Male to 1 and Female to 0 ( OneHotEncoding)
df['Sex'] = df['Sex'].map({'Male':1, 'Female':0})
```

```python
df.info()
```

#Splitting

```python
x = df.drop('Heart Attack Risk', axis=1)
y = df['Heart Attack Risk']
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, random_state= 0)
```

#Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)
```

#Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
regressor = LogisticRegression(penalty = 'l2', random_state = 0, max_iter= 100 )
regressor.fit(xtrain,ytrain)
ypred = regressor.predict(xtest)
```

#Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize the Random Forest Regressor with desired arguments
# You can tweak these arguments as per your requirements
random_forest_regressor = RandomForestRegressor(n_estimators=200, max_depth=30, random_state=42)

# Train the Random Forest Regressor
random_forest_regressor.fit(xtrain, ytrain)

# Predict on the testing data
ypred = random_forest_regressor.predict(xtest)
```

#Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
# Initialize the Gaussian Naive Bayes classifier
naive_bayes_classifier = GaussianNB()

# Train the classifier
naive_bayes_classifier.fit(xtrain, ytrain)
ypred = naive_bayes_classifier.predict(xtest)
```

#Calculating Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy_score(ypred,ytest)
```
